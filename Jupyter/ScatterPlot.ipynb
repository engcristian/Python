## Importing dependencies

```python
import pandas as pd 
import matplotlib.pyplot as pls 
pd.plotting.register_matplotlib_converters()
import seaborn as sns 
%matplotlib inline
```

## Importing the dataset 

```python
data_file = ('C:/Users/crist/Desktop/Python_Stuff/insurance.csv')
ins_charge = pd.read_csv(data_file)
print(ins_charge)
```

## sns.scatterplot()

```python
sns.scatterplot(x = ins_charge['bmi'], y= ins_charge['charges'])
```

## Regression Line with scatterplot sns.regplot()

```python
sns.regplot(x=ins_charge['bmi'], y= ins_charge['charges'])
```

##  Three variables with scatterplot

```python
sns.scatterplot(x = ins_charge['bmi'], y= ins_charge['charges'], hue= ins_charge['smoker'])
```

## Regression Linear with three variables using sns.lmplot()

```python
sns.lmplot(x = 'bmi', y ='charges', hue ='smoker', data= ins_charge)
```

## Adapting the design of the scatter plot to feature a categorical variable 'smoker'
## sns.swarmplot()

```python
sns.swarmplot(x = ins_charge['smoker'], y = ins_charge['charges'] )
```
