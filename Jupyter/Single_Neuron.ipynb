## Importing dependencies and loading the dataset

```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt 

# Set matplotlib 
plt.style.use('seaborn-whitegrid')
plt.rc('figure', autolayout= True)
plt.rc('axes', labelweight = 'bold', labelsize='large',titleweight = 'bold', titlesize = 18, titlepad = 10)

# Collecting data
data = 'C:/Users/crist/Desktop/Python_Stuff/red-wine.csv'
red_wine = pd.read_csv(data)
red_wine.head()
```

## Define the linear model
### units = 1 (quality) 
### input_shape = 11 (remaining columns are the features)

```python
linear_model = keras.Sequential([
    layers.Dense(units=1, input_shape = [11])
])
```

## Looking at the weights (w) and Bias (b)

```python
w, b = linear_model.weights
print(f'Weights:\n{w}\n\nBias:\n{b}')
```

## Plotting the output of an untrained linear model

```python
model = keras.Sequential([
    layers.Dense(1, input_shape = [1])
])

x = tf.linspace(-1.0, 1.0, 100)
y= model.predict(x)


plt.figure(dpi= 100)
plt.plot(x,y,'k')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel('Input: x')
plt.ylabel('Target: y')
w, b = model.weights
plt.title(f'Weight: {w[0][0]:0.2f}\nBias: {b[0]:0.2f}')
plt.show()
```
